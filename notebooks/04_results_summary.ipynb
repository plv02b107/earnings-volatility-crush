## Key Takeaways

4. Short volatility strategies suffer from:

## Final Conclusion

This study demonstrates that while earnings events are associated
with implied volatility collapse, **strategy design matters more than direction**.

Although iron condors are theoretically aligned with volatility crush,
their concave payoff structure exposes them to severe tail risk.

Conversely, long straddles benefit from convex payoffs
that dominate volatility decay during large earnings surprises.

**Conclusion:**  
Earnings trading is not simply about volatility levels,
but about **distributional asymmetry and payoff convexity**.
